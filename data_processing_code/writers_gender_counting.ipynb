```python
import pandas as pd
from pathlib import Path
```

```python
pd.set_option("display.max_rows", 20, "display.max_columns", None)
```

```python
genderpath = Path('./dialogue_writers_with_gender.csv')
genderdf = pd.read_csv(genderpath, encoding = 'utf-8')
```

```python
genderdf.head(10)
```

```python
# drop extra col indices col
genderdf.drop(genderdf.columns[0],axis=1,inplace=True)
```

```python
genderdf # 594 rows
```

```python
# convert gender col from string to list of strings
for i, row in genderdf.iterrows():
    gender_str = genderdf.at[i, 'writer_gender']
    gender_list = gender_str.split(',')
    genderdf.at[i, 'writer_gender'] = gender_list
```

```python
genderdf
```

```python
# count num of female writers in each movie
genderdf['num_of_female_writers'] = ''
```

```python
# see how many F writers each movie has
for i, row in genderdf.iterrows():
    gender_list = genderdf.at[i, 'writer_gender']
    f_count = gender_list.count('F')
    genderdf.at[i,'num_of_female_writers'] = f_count
```

```python
genderdf
```

```python
# add col for total num of writers
genderdf['total_num_of_writers'] = ''
for i, row in genderdf.iterrows():
    gender_list = genderdf.at[i, 'writer_gender']
    genderdf.at[i,'total_num_of_writers'] = len(gender_list)
```

```python
genderdf
```

```python
# calculate how many movies have atleast 1 female writer
atleast_one_f_writer = genderdf[genderdf['num_of_female_writers'] > 0]
atleast_one_f_writer
len(atleast_one_f_writer) # 99 out of 594
```

```python
# add col to genderdf to check if all writers for each movie are female
genderdf['all_female_writers'] = ''
```

```python
for i, row in genderdf.iterrows():
    gender_list = genderdf.at[i, 'writer_gender']
    if gender_list.count('M') == 0:
        genderdf.at[i, 'all_female_writers'] = True
    else:
        genderdf.at[i, 'all_female_writers'] = False
```

```python
# calculate how many movies have all female writers
all_female = genderdf[genderdf['all_female_writers']  == True]
all_female
len(all_female) # 22
```

```python
# value counts for num_of_female_writers
genderdf.num_of_female_writers.value_counts()
```

```python
# export complete writer gender data to csv
genderdf.to_csv("dialogue_writers_gender_with_counts.csv", encoding='utf-8')
```

# Visualizations

```python
piechart = genderdf.num_of_female_writers.value_counts().plot(kind = 'pie', 
                                                              figsize=(12,12), 
                                                              autopct = "%.1f%%", 
                                                              pctdistance = 0.7,
                                                              colormap = 'tab20',
                                                              textprops={'fontsize': 25})
                                                         
                                                  
piechart.set_title('Proportion of Movies with Female Writers', fontweight="bold", size=25)
labels = ['No female writers', '1 female writer', '2 female writers', '3 female writers']
piechart.legend(labels,loc = 'lower left', fontsize = 22)
piechart.set_ylabel('')
```
